```python
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report

import numpy as np
import os
import logging
import matplotlib.pyplot as plt
import seaborn as sns

# Cấu hình hệ thống Kaggle
BATCH_SIZE = 32 # [cite: 540, 597, 649]
IMG_SIZE = (224, 224) # Kích thước chuẩn cho EfficientNet [cite: 213, 641]
SEED = 42 # [cite: 543, 546]

# =====================
# ĐƯỜNG DẪN KAGGLE
# =====================
BASE_INPUT_PATH = "/kaggle/input/datasets/bananalatraichuoi/rambutan-uiters"
WORKING_PATH = "/kaggle/working/results"
os.makedirs(WORKING_PATH, exist_ok=True)

logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(message)s')
logger = logging.getLogger(__name__)

def run_reproduction_with_tensorflow():
    # 1. Tải và tiền xử lý dữ liệu bằng TensorFlow
    # Sử dụng image_dataset_from_directory thay cho ImageFolder của PyTorch
    full_dataset = tf.keras.utils.image_dataset_from_directory(
        BASE_INPUT_PATH,
        image_size=IMG_SIZE,
        batch_size=None, # Tải ảnh đơn lẻ để thực hiện split thủ công
        label_mode='categorical', # [cite: 539]
        shuffle=True,
        seed=SEED
    )

    # Chuyển đổi sang list để thực hiện split 80:20 [cite: 9, 255, 550]
    images = []
    labels = []
    for img, label in full_dataset:
        images.append(img.numpy())
        labels.append(label.numpy())
    
    images = np.array(images)
    labels = np.array(labels)
    
    X_train, X_test, y_train, y_test = train_test_split(
        images, labels, test_size=0.2, random_state=SEED, stratify = np.argmax(labels, axis=1)
    )

    # Chia thêm tập Validation từ tập Train (10% của 80%) [cite: 258, 551]
    X_train, X_val, y_train, y_val = train_test_split(
        X_train, y_train, test_size=0.1, random_state=SEED
    )

    logger.info(f"📊 Train: {len(X_train)} | Val: {len(X_val)} | Test: {len(X_test)}")

    # 2. Xây dựng Data Pipeline với Data Augmentation [cite: 45, 214, 642]
    data_augmentation = tf.keras.Sequential([
        layers.RandomRotation(0.05), # Khoảng 20 độ [cite: 213]
        layers.RandomFlip("horizontal_and_vertical"), # [cite: 213, 552]
        layers.RandomZoom(0.2), # [cite: 213]
    ])

    # 3. Khởi tạo mô hình EfficientNet-B0 [cite: 10, 46, 204]
    base_model = tf.keras.applications.EfficientNetB0(
        weights='imagenet', # Sử dụng trọng số pre-trained [cite: 87, 614]
        include_top=False, 
        input_shape=(224, 224, 3)
    )

    # Chiến lược đóng băng (Fine-tuning): Đóng băng các lớp đầu [cite: 619, 651]
    # base_model.trainable = True
    # for layer in base_model.layers[:100]: 
    #     layer.trainable = False

    # 4. Thêm các lớp phân loại tùy chỉnh [cite: 270, 442]
    num_classes = labels.shape[1]
    model = models.Sequential([
        layers.Input(shape=(224, 224, 3)),
        data_augmentation,
        base_model,
        layers.GlobalAveragePooling2D(), # [cite: 394, 484]
        layers.Dropout(0.3),             # [cite: 610]
        layers.Dense(num_classes, activation='softmax') # [cite: 310, 487]
    ])

    # 5. Cấu hình huấn luyện [cite: 260, 539, 648]
    # model.compile(
    #     optimizer=optimizers.Adam(learning_rate=0.0001), # [cite: 591, 648]
    #     loss='categorical_crossentropy', # [cite: 539]
    #     metrics=['accuracy']
    # )

    # # 6. Callbacks để bảo vệ quá trình huấn luyện [cite: 540, 652]
    # callbacks = [
    #     tf.keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True), # [cite: 540, 624-627]
    #     tf.keras.callbacks.ReduceLROnPlateau(factor=0.2, patience=3) # [cite: 652]
    # ]

    # # 7. Bắt đầu huấn luyện [cite: 601-605]
    # logger.info("🚀 Bắt đầu huấn luyện...")
    # history = model.fit(
    #     X_train, y_train,
    #     validation_data=(X_val, y_val),
    #     epochs=30, # [cite: 601-605]
    #     batch_size=BATCH_SIZE, # [cite: 540, 597, 649]
    #     callbacks=callbacks
    # )

    # Stage 1
    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-3),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=20)
    
    # Stage 2 Fine-tune
    base_model.trainable = True
    for layer in base_model.layers[:-20]:
        layer.trainable = False
    
    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-5),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    history = model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=15)

    class_names = list(full_dataset.class_names)
    evaluate_detailed_performance(model, X_test, y_test, class_names)
    # Đánh giá trên tập Test
    test_loss, test_acc = model.evaluate(X_test, y_test)
    logger.info(f"✅ Kết quả Test - Accuracy: {test_acc*100:.2f}%")
    
def evaluate_detailed_performance(model, X_test, y_test, class_names):
    # 1. Dự đoán kết quả trên tập Test
    logger.info("🔮 Đang dự đoán trên tập dữ liệu kiểm thử...")
    y_pred_probs = model.predict(X_test)
    y_pred = np.argmax(y_pred_probs, axis=1)
    y_true = np.argmax(y_test, axis=1)

    # 2. Tạo Báo cáo phân loại (Accuracy, Precision, Recall cho từng class) [cite: 1561, 1565, 1575]
    # Đây là nơi bạn xem được "Accuracy theo class" thông qua chỉ số F1-score và Recall [cite: 542]
    report = classification_report(y_true, y_pred, target_names=class_names)
    print("\n📊 BÁO CÁO PHÂN LOẠI CHI TIẾT (CLASSIFICATION REPORT):")
    print(report)

    # 3. Tính toán Ma trận nhầm lẫn (Confusion Matrix) 
    cm = confusion_matrix(y_true, y_pred)
    
    # 4. Vẽ Ma trận nhầm lẫn bằng Seaborn
    plt.figure(figsize=(12, 10))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=class_names, yticklabels=class_names)
    plt.title('Ma trận nhầm lẫn (Confusion Matrix) - Dự án Chôm chôm/Ớt') # [cite: 48, 1205]
    plt.ylabel('Thực tế (True Label)')
    plt.xlabel('Dự đoán (Predicted Label)')
    
    # Lưu biểu đồ vào thư mục làm việc của Kaggle
    plt.savefig(os.path.join(WORKING_PATH, 'confusion_matrix.png'))
    plt.show()
    logger.info(f"✅ Đã lưu ma trận nhầm lẫn tại {WORKING_PATH}")

if __name__ == "__main__":
    run_reproduction_with_tensorflow()
```
